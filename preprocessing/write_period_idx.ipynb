```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nrows = None
```

```python
input_dir = '/run/media/hoosiki/WareHouse3/mtb/datasets/LANL'
```

```python
df_period = pd.DataFrame(columns=['period', 'last_idx', 'time'])
```

```python
df_train = pd.read_csv(input_dir + '/train.csv', nrows=nrows)
df_train
```

```python
period = 0
remaining_time = 99999999
for idx, time in enumerate(df_train['time_to_failure']):
    if remaining_time < time:
        print(period, " ", idx - 1, " ", remaining_time)
        df_period = df_period.append({'period': period, 'last_idx': idx - 1, 'time': remaining_time}, ignore_index=True)
        period += 1
    remaining_time = time    
```

```python
df_period['period'] = df_period['period'].astype(int)
df_period['last_idx'] = df_period['last_idx'].astype(int)
df_period
```

```python
df_period.to_csv('period2idx.csv')
```
